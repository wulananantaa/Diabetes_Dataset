```python
import zipfile

with zipfile.ZipFile("pima-indians-diabetes-database.zip", 'r') as zip_ref:
    zip_ref.extractall("pima_data")
```

```python
import pandas as pd

df = pd.read_csv("pima_data/diabetes.csv")
df.head()
```

```python
# Informasi struktur data
print(df.info())

# Statistik deskriptif
print(df.describe())

# Cek nilai unik pada target
print(df['Outcome'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Korelasi antar fitur
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Distribusi target
sns.countplot(data=df, x='Outcome')
plt.title("Distribusi Diabetes (0 = Tidak, 1 = Ya)")
plt.show()
```
